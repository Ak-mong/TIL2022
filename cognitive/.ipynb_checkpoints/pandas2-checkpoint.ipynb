```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
births   #read_csv 가 function(함수?)
```

# 피벗 테이블

```python
births['decade'] = births['year']//10*10  # // : 버림  # decade를 맨앞으로 옮길수없을까?
births
```

```python
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum')
# pivot_table안의 births는 변수 births가 아니라 표안의 births
```

```python
import matplotlib.pyplot as plt
```

```python
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum').plot()
```
